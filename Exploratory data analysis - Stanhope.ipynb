```python
clean_df = new_house_df.drop(['Property_type','Address','Agent_Names','Agency_Name','url','Date Sold'],axis=1)
```

```python
clean_df.head()
```

```python
X = clean_df[['Property_Area','Beds','Bath','Parking']]
y = clean_df['Price_as_float']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

1 sq meter of property in stanhope costs 647 dollars
1 Each bedroom approx. accounts for 52k dollars
1 bath room about 100k (????)
1 parking about 35K

```python
predictions = lm.predict(X_test)
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(y_test,predictions)
```

```python
plt.figure(figsize=(8,8))
sns.distplot((y_test-predictions),bins=50)b
```

Majority of the variations is around the zero, indicating relatively ok predictions ?

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

### RMSE is 100k, predictions to vary 100k approx for any given property

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,predictions)
```


### This simple model can explain 57% of the variation of data... not bad for a simple model, though more feature engineering needs to be done.
